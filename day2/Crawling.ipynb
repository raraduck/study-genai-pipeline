- 헤드라인 뉴스 10개 가져와 제목/요약/url(링크) 추출하기
- pandas 형식으로 출력

```python
import requests
from bs4 import BeautifulSoup as bs

html = requests.get('https://news.naver.com/section/102')

soup = bs(html.text, 'html.parser')
find_ul = soup.find('ul', class_="sa_list")
print(find_ul)
print(len)
```

```python
# 헤드라인 뉴스 정보 가져오기
div_texts = find_ul.find_all('div', class_="sa_text")
len(div_texts)
```

```python
# 각각의 정보를 추출해 Pandas 형식으로 출력
import pandas as pd

divTexts = [] # 데이터 저장할 리스트 생성

for el in div_texts:
    title = el.find('a').get_text
    linkUrl = el.find('a')['href']
    doc = el.find('div', class_="sa_text_lede").get_text()
    divTexts.append({
        '제목': title,
        '요약': doc,
        '원본(링크)': linkUrl
        })

df = pd.DataFrame(divTexts)
df
```
