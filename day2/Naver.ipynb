```python
from bs4 import BeautifulSoup as bs
import requests
import sys
import pandas as pd
```

```python
url = 'https://kin.naver.com/search/list.naver?query=%EC%A0%9C%EC%A3%BC%EB%8F%84&page='
html = requests.get(url + str(1))
soup = bs(html.text)
# print(soup)

ul_soup = soup.find('ul', class_='basic1')
li_soup = ul_soup.find_all('li')
# li_soup[0]

df_list = []
for item in li_soup:
    title = item.find('dt').get_text()
    link = item.find('dt').find('a')['href']
    dd_soup = item.find_all('dd') # 모든 <dd> 태그 검색
    indate = dd_soup[0].get_text() # 입력일
    field = dd_soup[1].get_text() # 요약
    tNum = dd_soup[-1].find('span', class_="hit").get_text().split(' ')[-1] # 답변수

    df_list.append({
        '제목': title, 
        '내용': field,
        '입력일': indate,
        '답변수': tNum,
        '원본(링크)': link
    })

df = pd.DataFrame(df_list)
df
```
